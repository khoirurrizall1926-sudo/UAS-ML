```python
# --- 1. IMPORT LIBRARY ---
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.datasets import load_breast_cancer # Hanya untuk mengambil nama kolom yang cantik

# --- 2. LOAD DATASET ---
filename = 'breast_cancer.csv'

try:
    # Skip baris 1 (metadata), header=None
    df = pd.read_csv(filename, header=None, skiprows=1)

    # Memberi nama kolom agar rapi
    nama_fitur_resmi = load_breast_cancer().feature_names
    df.columns = list(nama_fitur_resmi) + ['Diagnosis']

    print("✅ Dataset berhasil dimuat!")

except FileNotFoundError:
    print("❌ File tidak ditemukan. Pastikan sudah upload 'breast_cancer.csv'")
    raise

    from sklearn.datasets import load_breast_cancer

# 1. Mengambil nama fitur resmi dari sklearn
# (Karena file CSV 'breast_cancer.csv' tidak memiliki header/judul kolom)
nama_fitur_resmi = load_breast_cancer().feature_names

# 2. Menampilkan nama fitur ke layar
print("Daftar Nama Fitur:")
print(nama_fitur_resmi)

# Pisahkan Fitur (X) dan Target (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]


# Hitung jumlah per kelas
counts = y.value_counts()


print("Distribusi Kelas Target:")
# Loop untuk menampilkan hasil hitungan
for label, jumlah in counts.items():
    print(f" - Kelas {label} : {jumlah} data")


# --- 3. MENAMPILKAN INFO DATASET ---
print("\n" + "="*40)
print("       INFO DATASET")
print("="*40)
print(f"Jumlah Data Total : {X.shape[0]}")
print(f"Jumlah Fitur      : {X.shape[1]}")
print(f"Distribusi Target :")
print(y.value_counts().to_string())

# --- 4. SPLIT DATA (70% Train - 30% Test) ---
# test_size=0.3 artinya 30% untuk testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("\n[INFO SPLIT DATA]")
print(f"Data Training : {X_train.shape[0]} baris")
print(f"Data Testing  : {X_test.shape[0]} baris")

# --- 5. FEATURE SCALING (STANDARD SCALER) ---
# Penting: KNN menghitung jarak, jadi skala data harus disamakan dulu
scaler = StandardScaler()

# Fit hanya pada data TRAIN, lalu transform ke train dan test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- 6. MODELING KNN (K=5) ---
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train_scaled, y_train)

# --- 7. PREDIKSI ---
y_pred = knn.predict(X_test_scaled)

# --- 8. EVALUASI METRICS ---
print("\n" + "="*40)
print(f"    HASIL EVALUASI (KNN k={k})")
print("="*40)

# Menghitung metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"1. Accuracy  : {acc:.4f} ({acc*100:.2f}%)")
print(f"2. Precision : {prec:.4f}")
print(f"3. Recall    : {rec:.4f}")
print(f"4. F1-Score  : {f1:.4f}")

print("\nLaporan Detail:")
print(classification_report(y_test, y_pred, target_names=['Malignant', 'Benign']))
```
